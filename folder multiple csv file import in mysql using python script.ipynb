```python
import os
import pandas as pd
from sqlalchemy import create_engine

# MySQL connection string
engine = create_engine("mysql+mysqlconnector://root:012345@127.0.0.1:3306/sales_performance_analytics_db")

folder_path = r"C:\Users\Datchnamoorthy R\OneDrive\Desktop\Project work 2026\mysql + power bi project\dataset"

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)

        # Table name = file name without extension
        table_name = os.path.splitext(file)[0]

        # Create/replace table for each CSV
        df.to_sql(table_name, con=engine, if_exists="replace", index=False)
        print(f"{file} imported successfully into table {table_name}!")
```
